```python
import cv2
import os
import shutil
```

```python

# Set the input and output folder paths
input_dir = r"/media/arush/New Volume/Work/Giloma_Kaggle/Data/Glioma_MDC_2025_training/Glioma_MDC_2025_training"
output_dir = r"/media/arush/New Volume/Work/Giloma_Kaggle/Data/upsampled/training"

# If the output folder does not exist, create it
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Target size (1015x1015)
target_size = (1015, 1015)

# Get all .jpg image files from the input folder
image_files = [f for f in os.listdir(input_dir) if f.lower().endswith('.jpg')]

# # Batch upsampling
for img_name in image_files:
    input_path = os.path.join(input_dir, img_name)
    output_path = os.path.join(output_dir, img_name)

    # Read the original image
    img = cv2.imread(input_path)

    if img is None:
        print(f"Unable to read image: {input_path}, skipping...")
        continue

    # Use bicubic interpolation for upsampling
    resized_img = cv2.resize(img, target_size, interpolation=cv2.INTER_CUBIC)

    # Save the processed image
    cv2.imwrite(output_path, resized_img)
    print(f"Processing completed: {output_path}")
```

```python

json_files = [f for f in os.listdir(input_dir) if f.lower().endswith('.json')]

# Copy each .json file to the output folder
for json_name in json_files:
    json_input_path = os.path.join(input_dir, json_name)
    json_output_path = os.path.join(output_dir, json_name)
    
    shutil.copy(json_input_path, json_output_path)
    print(f"Copied: {json_output_path}")
```
